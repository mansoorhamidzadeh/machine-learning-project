```python
import numpy as np
import  pandas as pd
import  matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler,MinMaxScaler
```

```python
advertising=pd.read_csv('Advertising.csv')
advertising.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
sns.heatmap(advertising.corr(),annot=True,cmap='gray')
```

```python
x=np.reshape(advertising['TV'].values,(-1,1))
y=np.reshape(advertising['Sales'].values,(-1,1))
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.7)
```

```python
lin=LinearRegression()
```

```python
lin.fit(x_train,y_train)
```

```python
y_pred=lin.predict(x_test)
```

```python
print(metrics.r2_score(y_test,y_pred))
```

```python
res=(y_test-y_pred)
```

```python
res
```

```python

sns.displot(res,bins=20)
```

```python
print(metrics.mean_squared_error(y_test,y_pred))
```

```python
r2=metrics.r2_score(y_test,y_pred)
r2
```

```python
plt.scatter(x_test,y_test)
plt.plot(x_test, lin.intercept_ + lin.coef_*x_test, 'r')
```
