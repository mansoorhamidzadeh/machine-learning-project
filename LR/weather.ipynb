```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error,r2_score
from sklearn.linear_model import LinearRegression
```

```python
data=pd.read_csv('ww-ii-data.csv')
data
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
x=data[['STA','YR','MO','DA']]
y=data[['MaxTemp']]
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=3243)
```

```python

li_model=LinearRegression()
li_model.fit(x_train,y_train)
x_test
```

```python
li_model.predict([x_test.iloc[100]])
```

```python
y_pred=li_model.predict(x_test)
```

```python
mean_absolute_error(y_test,y_pred)
```

```python
r2_score(y_test,y_pred)*100
```

```python
data=data.fillna(data.mean())
```

```python
data
```

```python
data=data.dropna(axis=1)
```

```python
data.isna().sum()
```

```python
x=data[['STA','WindGustSpd','YR','MO','DA','SPD','MAX','MIN','MEA','SPD','PGT','MeanTemp','MinTemp']]
y=data[['MaxTemp']]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=3243)
li_model=LinearRegression()
li_model.fit(x_train,y_train)
```

```python
y_pred=li_model.predict(x_test)
```

```python
print(mean_absolute_error(y_test,y_pred))
```

```python
r2_score(y_test,y_pred)*100
```
